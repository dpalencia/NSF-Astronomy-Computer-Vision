```python
from marsvision.pipeline.Model import Model
from marsvision.utilities.DataLoader import DataLoader
from sklearn.ensemble import RandomForestClassifier
import os
import matplotlib.pyplot as plt
import cv2
```

```python
# Load in data using our dataloader
loader = DataLoader("../../mars-dataset")
loader.data_reader()
loader.data_transformer()
data = loader.df
rf = RandomForestClassifier(n_estimators = 500)
data.head()
```

```python
# Set class labels (boolean 1 == true, 0 == false)
labels = [1 if name == "dust" else 0 for name in loader.labels]
```

```python
# Pass an sklearn model into the model class and train it
rf = RandomForestClassifier(n_estimators = 500)
model = Model(rf, "sklearn", training_images = loader.images, training_labels = labels)
```

```python
model.train_model()
```

```python
# Run a sliding window across the image, 
# make predictions on each segment,
# show the image segments and their predictions using subplots

def plot_windows(window_length, window_height, test_image):
    fig=plt.figure(figsize=(8, 8))
    columns = int(test_image.shape[0] / window_length)
    rows = int(test_image.shape[1] / window_height)
    
    # Windows to show with subplots
    windows = []
    # Model predictions
    classes = []
    
    for y in range(0, test_image.shape[0], window_length):
        for x in range(0, test_image.shape[1], window_height):
            # the window algorithm should handle the case 
            # where the remainder of the image in one axis is smaller
            # than the window's size on that axis
            
            # Additionally, we should be able to specify the stride
            # of the window here it's assumed that the stride)
            # is the same size as the window
            window = test_image[x:window_length + x + 1, y:window_height + y + 1]
            windows.append(window)
            if(window.size > 0):
                prediction = model.predict(window)
                classes.append(prediction)

    for i  in range(rows * columns):
        ax = fig.add_subplot(rows, columns, i + 1)
        ax.title.set_text(str(classes[i]))
        plt.imshow(windows[i])
        plt.axis("off")
    plt.show()
```

```python
plot_windows(window_length, window_height, loader.images[5])
```

```python
plot_windows(window_length, window_height, loader.images[250])
```
